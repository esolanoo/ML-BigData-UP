### Ejercicio 1

Para mostrar los productos más caros de cada marca, también el promedio de los precios.



```python
query  =  """
SELECT m.title AS "Brand", max(a.price) AS max_price, avg(a.price) AS avg_price 
FROM articles a
JOIN manufacturers m ON m.id=a.manufacturer_id
GROUP BY m.id
LIMIT 10
"""

run(query)
```

### Ejercicio 2

Haz un query para mostrar cuántos productos existen en cada categoría.

```python
query = """
SELECT m.title as "Brand", count(a.id) AS "Article count"
FROM articles a
JOIN manufacturers m ON m.id=a.manufacturer_id
GROUP BY m.id
LIMIT 10
"""

run(query)
```

### Ejercicio 4

Muestra los 5 productos más relevantes de cada marca.

```python
query = """
SELECT
    Brand,
    GROUP_CONCAT(Product SEPARATOR ', ') AS Products
FROM (
    SELECT
        a.title AS "Product",
        m.title AS "Brand",
        ROW_NUMBER() OVER (PARTITION BY m.id ORDER BY a.price DESC) AS rn
    FROM
        articles a
        JOIN manufacturers m ON m.id = a.manufacturer_id
) AS ranked_products
WHERE
    rn <= 5
GROUP BY
    Brand;

"""

run(query).tail(2).to_markdown()
```
